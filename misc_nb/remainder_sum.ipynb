```python
import qt
from qt import dt, np, pd
```

```python
def f(n):
	rs = n% (1 + np.arange(n))
	return rs
	# return sum(rs)
```

```python
f(99)
f(100)
```

```python
r_99 = f(99)
r_100 = f(100)
```

```python
t = pd.DataFrame()
t['div'] = 1 + np.arange(100)
t['r_99'] = np.append(r_99, 0)
t['r_100'] = r_100
```

```python
qt.view(t)
```

```python
t_div = t[(t['r_99'] == 0) | (t['r_100'] == 0)]
t_div['r'] = (t_div['r_99'] - t_div['r_100']) 
qt.view(t_div)
```

```python
t = pd.read_csv('spy_history.csv')
t['date'] = pd.to_datetime(t['date'])
t = t.sort_values('date').reset_index(drop=True)
t['ret_1d'] = 100*t['close'].pct_change().fillna(0)
```

```python
for h in [20, 60, 120, 250]:
	t[f'var_{h}d'] = t['ret_1d'].rolling(window=h).var()
```

```python
t['var_20d'].mean()
t['var_60d'].mean()
t['var_120d'].mean()
t['var_250d'].mean()
```

```python
t.plot(x='date', y=[f'var_{h}d' for h in [20, 60, 120, 250]])
```
