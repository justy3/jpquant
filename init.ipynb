#### get universe

```python
def get_universe_asof_date(date=dt.date.today().strftime("%Y%m%d"), code=""):

	qt.log.info(f"querying universe as of date = {date}, code = {code}")

	# create parameters to send universe request with
	params = {}
	if code != "":
		params["code"] = code
	if date != "":
		params["date"] = date

	# send request
	res = requests.get(f"{JP_API_URL}/v1/listed/info", params=params, headers=headers)

	# dummy result
	df = pd.DataFrame()

	# if request is successful, get all data with pagination
	if res.status_code == 200:
		d = res.json()
		data = d["info"]
		while "pagination_key" in d:
			params["pagination_key"] = d["pagination_key"]
			res = requests.get(f"{JP_API_URL}/v1/listed/info", params=params, headers=headers)
			d = res.json()
			data += d["info"]
		
		# create dataframe
		df = pd.DataFrame(data)
		cs = df.columns.tolist()
		df['query_date'] = pd.to_datetime(date).date()
		df = df[['query_date'] + cs]
		df['Date'] = pd.to_datetime(df['Date']).dt.date

	else:
		qt.log.error(f"failed to get universe on date = {date}: {res.status_code} {res.text}")

	return df
```

```python
TARGET_SYMBOLS = [
	"7201",  # 日産
	"7203",  # トヨタ
	"7267",  # ホンダ
]
BASE_DATE = "2020-01-01"
TARGET_SYMBOLS  = univ[univ['ScaleCategory'].isin(["TOPIX Core30", "TOPIX Large70"])]['Code'].unique().tolist()
len(TARGET_SYMBOLS)
```

```python
buff = []
for code in TARGET_SYMBOLS:
	df = jqapi.get_prices_daily_quotes(code=code)
	buff.append(df)
df_p = pd.concat(buff).reset_index(drop=True)

# 各列のデータ型を調整します
df_p.loc[:, "Date"] = pd.to_datetime(df_p["Date"], format="%Y-%m-%d")
df_p.loc[:, "Open"] = df_p["Open"].astype(np.float64)
df_p.loc[:, "High"] = df_p["High"].astype(np.float64)
df_p.loc[:, "Low"] = df_p["Low"].astype(np.float64)
df_p.loc[:, "Close"] = df_p["Close"].astype(np.float64)
df_p.loc[:, "Volume"] = df_p["Volume"].astype(np.float64)
df_p.loc[:, "TurnoverValue"] = df_p["TurnoverValue"].astype(np.float64)
df_p.loc[:, "AdjustmentFactor"] = df_p["AdjustmentFactor"].astype(np.float64)
df_p.loc[:, "AdjustmentOpen"] = df_p["AdjustmentOpen"].astype(np.float64)
df_p.loc[:, "AdjustmentHigh"] = df_p["AdjustmentHigh"].astype(np.float64)
df_p.loc[:, "AdjustmentLow"] = df_p["AdjustmentLow"].astype(np.float64)
df_p.loc[:, "AdjustmentClose"] = df_p["AdjustmentClose"].astype(np.float64)
df_p.loc[:, "AdjustmentVolume"] = df_p["AdjustmentVolume"].astype(np.float64)

# 日付を絞り込みます
df_p = df_p.loc[df_p["Date"] >= BASE_DATE]
```

```python
df_p.to_csv("jpquant_prices_daily_quotes.csv", index=False)
```
